```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('la_nugs.csv')
```

```python
top_ = df.head(10)
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
df1 = df.drop(['Opponent'],axis=1)
```

```python
scaler = StandardScaler()
scaler.fit(df1)
```

```python
scaled_data=scaler.transform(df1)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(scaled_data)
```

```python
x_pca=pca.transform(scaled_data)
```

```python
scaled_data.shape
```

```python
x_pca.shape
```

```python
plt.figure(figsize=(10,10))
plt.scatter(x_pca[:,0],x_pca[:,1],c=df['Team'])
plt.xlabel("first pca")
plt.ylabel("2nd pca")
```
